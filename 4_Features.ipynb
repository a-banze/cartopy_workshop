## Adding features (Natural Earth)
Demonstration of the built-in Natural Earth features found in Cartopy.

```python
import matplotlib.pyplot as plt

import cartopy.crs as ccrs
import cartopy.feature as cfeature


fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.set_extent([-20, 60, -40, 45], crs=ccrs.PlateCarree())

ax.add_feature(cfeature.LAND)  # Land polygons, including major islands
ax.add_feature(cfeature.OCEAN)  # Ocean polygons
ax.add_feature(cfeature.COASTLINE)  # Coastline, including major islands
ax.add_feature(cfeature.BORDERS, linestyle=':')  # Country boundaries
ax.add_feature(cfeature.LAKES, alpha=0.5)  # Natural and artificial lakes
ax.add_feature(cfeature.RIVERS)  # Single-line drainages, including lake centerlines

plt.show()
```
